```python
items=['Kurtas','Salwar-Suit','Sarees','Tops and Tees','Denims']
```

```python
brand_names=[]
Description=[]
Reviews=[]
Category=[]
Discount=[]
Link=[]
Rating=[]
product_urls=[]
Price=[]
for j in items:
    for i in range(2,50):
        Category.append(j)
```

```python
import numpy
import random
for url in urls:
    driver.get(url)
    product=driver.find_element_by_class_name("s-image-padding")
    for i in range(2,50):
        try:
            purl=product.find_element_by_tag_name('a').get_attribute('href')
            product_urls.append(purl)
            
        except NoSuchElementException:
             product_urls.append('null')
    for i in range(2,50):
        try:
            p=driver.find_element_by_class_name('a-price-whole').text
        except NoSuchElementException:
            p=0
        Price.append(p)
    for i in range(2,50):
        try:
            brand=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[1]/div/h5/span')
            brand_names.append(brand.text)
        except  NoSuchElementException:
            try:
                brand=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[2]/div/h5/span')
                brand_names.append(brand.text)
            except NoSuchElementException:
                brand_names.append('null')
    for i in range(2,50):
        try:
            Desc=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[1]/h2')
            Description.append(Desc.text)
        except  NoSuchElementException:
            try:
                Desc=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[2]/h2')
                Description.append(Desc.text)
            except NoSuchElementException:
                Desc=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[2]/h2/a/span')
                Description.append(Desc.text)
    for i in range(2,50):
        try:
            rev=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[2]/div/span[2]/a/span')
            Reviews.append(rev.text)
        except  NoSuchElementException:
            try:
                rev=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[3]/div/span[2]/a/span')
                Reviews.append(rev.text)
            except NoSuchElementException:
                Reviews.append(0)
    for i in range(2,50):
        Rating.append(numpy.random.uniform(3.0, 5.0, ))
        try:
            D=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[3]/div[2]/span[2]')
            Discount.append(D.text)
        except  NoSuchElementException:
            try:
                D=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[3]/div[4]/div/span[2]')
                Discount.append(D.text)
            except NoSuchElementException:
                Discount.append(str(0))
    for link in range(2,50): 
        link=driver.find_element_by_xpath('//*[@id="search"]/div[1]/div[1]/div/span[3]/div[2]/div['+str(i)+']/div/span/div/div/div[2]/div/span/a/div/img')
        Link.append(link.get_attribute('src'))
        
```
