# Reading data

```python
import pandas as pd
```

```python
dfM = pd.read_csv("myntra_men_latest.csv")
print("shape: ", dfM.shape)
dfM.head()
```

```python
dfF = pd.read_csv("myntra_women_latest.csv")
print("shape: ", dfF.shape)
dfF.head()
```

```python
df = pd.concat([dfM, dfF])
df.shape
```

# Pre-Processing

```python
# Removing Varified Buyers from reviews
df["reviews"] = df["reviews"].str.split(expand=True)[0]
df
```

```python
df.reviews = (df.reviews.replace(r'[Kk]+$', '', regex=True).astype(float) * df.reviews.str.extract(r'[\d\.]+([Kk]+)', expand=False).fillna(1).replace(['K','k'], [10**3, 10**3]).astype(int))
df
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mean())
df["reviews"] = df["reviews"].fillna(df["reviews"].mean())
df
```
